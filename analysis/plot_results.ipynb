```python
import pandas as pd
import matplotlib.pyplot as plt

# List of (experiment names, label names)

## these are deprecated:
#     ("PegTransfer-targetblock-10000demo-1e5_0", "Target block position (10K)") # 50 epochs
#     ("PegTransfer-14_targetblocktargetpeg-1000demo-1e5_0", "Target block and goal peg positions"),
#     ("PegTransfer-targetblock-1000demo-1e5_0", "Target block position"),
    ## TWO BLOCKS - done
#     ('PegTransferTwoBlocksNoColor-onehot-1000demo-2e5_0_r0', 'PegTransferTwoBlocksNoColor-onehot-1000demo-2e5_0_r1', 'PegTransferTwoBlocksNoColor-onehot-1000demo-2e5_0_r2', 'Block encoding'),
#     ('PegTransferTwoBlocksNoColor-onehottargetpeg-1000demo-2e5_0_r0', 'PegTransferTwoBlocksNoColor-onehottargetpeg-1000demo-2e5_0_r1', 'PegTransferTwoBlocksNoColor-onehottargetpeg-1000demo-2e5_0_r2', 'Block encoding and goal position'),
#     ('PegTransferTwoBlocksNoColor-fourtuple-1000demo-2e5_0_r0', 'PegTransferTwoBlocksNoColor-fourtuple-1000demo-2e5_0_r1', 'PegTransferTwoBlocksNoColor-fourtuple-1000demo-2e5_0_r2', 'Color encoding'), 
#     ('PegTransferTwoBlocksNoColor-targetblock-1000demo-2e5_0_r0', 'PegTransferTwoBlocksNoColor-targetblock-1000demo-2e5_0_r1', 'PegTransferTwoBlocksNoColor-targetblock-1000demo-2e5_0_r2', 'Target block position'),
#     ('PegTransferTwoBlocksNoColor-targetblocktargetpeg-1000demo-2e5_0_r0', 'PegTransferTwoBlocksNoColor-targetblocktargetpeg-1000demo-2e5_0_r1', 'PegTransferTwoBlocksNoColor-targetblocktargetpeg-1000demo-2e5_0_r2', 'Target peg and goal positions') 
    
    ## TWO BLOCKS COLOR - done
#     ('PegTransferTwoBlocks-onehot-1000demo-2e5_0_r0', 'PegTransferTwoBlocks-onehot-1000demo-2e5_0_r1', 'PegTransferTwoBlocks-onehot-1000demo-2e5_0_r2', 'Block encoding'),
#     ('PegTransferTwoBlocks-fourtuple-1000demo-2e5_0_r0', 'PegTransferTwoBlocks-fourtuple-1000demo-2e5_0_r1', 'PegTransferTwoBlocks-fourtuple-1000demo-2e5_0_r2', "Color encoding"),
#     ('PegTransferTwoBlocks-targetblock-1000demo-2e5_0_d1', 'PegTransferTwoBlocks-targetblock-1000demo-2e5_0_d2', 'PegTransferTwoBlocks-targetblock-1000demo-2e5_0_d3', 'Target block position'),
#     ('PegTransferTwoBlocks-targetblocktargetpeg-1000demo-2e5_0_d1', 'PegTransferTwoBlocks-targetblocktargetpeg-1000demo-2e5_0_d2', 'PegTransferTwoBlocks-targetblocktargetpeg-1000demo-2e5_0_d3', 'Target block and goal peg positions'),
#     ('PegTransferTwoBlocks-onehottargetpeg-1000demo-2e5_0_d1', 'PegTransferTwoBlocks-onehottargetpeg-1000demo-2e5_0_d2', 'PegTransferTwoBlocks-onehottargetpeg-1000demo-2e5_0_d3', 'Block encoding and goal position'),
"""
- Goal block position
- ⁠RGBA color
- ⁠One hot block encoding
- ⁠Target peg position"""
    
experiment_names = [
    ## ALL BLOCKS - done
#     ("PegTransfer-onehot-1000demo-2e5_0", 'PegTransfer-onehot-1000demo-2e5_0_r1', 'PegTransfer-onehot-1000demo-2e5_0-d1', "⁠One hot block encoding"),
#     ("PegTransfer-14_onehottargetpeg-1000demo-2e5_0",'PegTransfer-14_onehottargetpeg-1000demo-2e5_0-d1','PegTransfer-14_onehottargetpeg-1000demo-2e5_0-d2',"⁠One hot block encoding and target peg position"),
#     ('PegTransfer-fourtuple-1000demo-2e5_0_r0', 'PegTransfer-fourtuple-1000demo-2e5_0_r1', 'PegTransfer-fourtuple-1000demo-2e5_0_r2', '⁠RGBA color'),
#     ("PegTransfer-targetblock-1000demo-2e5_0", 'PegTransfer-targetblock-1000demo-2e5_0_r1', 'PegTransfer-targetblock-1000demo-2e5_0_r2', "Goal block position"),
#     ("PegTransfer-14_targetblocktargetpeg-1000demo-2e5_0", 'PegTransfer-14_targetblocktargetpeg-1000demo-2e5_0_r1', 'PegTransfer-14_targetblocktargetpeg-1000demo-2e5_0_r2', "Goal block and target peg positions"),
    
    ## COLOR - done
#     ("PegTransferColor-onehot-1000demo-2e5_0","PegTransferColor-onehot-1000demo-2e5_0-d1","PegTransferColor-onehot-1000demo-2e5_0-d2", "⁠One hot block encoding"), #color
#     ("PegTransferColor-14_onehottargetpeg-1000demo-2e5_0", "PegTransferColor-14_onehottargetpeg-1000demo-2e5_0_r1","PegTransferColor-14_onehottargetpeg-1000demo-2e5_0_r2", "⁠One hot block encoding and target peg position"), #color
#     ('PegTransferColor-fourtuple-1000demo-2e5_0_r0', 'PegTransferColor-fourtuple-1000demo-2e5_0_r1', 'PegTransferColor-fourtuple-1000demo-2e5_0_r2', '⁠RGBA color'), # color
#     ("PegTransferColor-targetblock-1000demo-2e5_0", 'PegTransferColor-targetblock-1000demo-2e5_0-d1', 'PegTransferColor-targetblock-1000demo-2e5_0-d2',"Goal block position"), #color
#     ("PegTransferColor-14_targetblocktargetpeg-1000demo-2e5_0", 'PegTransferColor-14_targetblocktargetpeg-1000demo-2e5_0_r1', 'PegTransferColor-14_targetblocktargetpeg-1000demo-2e5_0_r2', "Goal block and target peg positions"), #color

    
    ## Target block ONLY
#     ('PegTransferTBOTwoBlock-1000demo-2e5_0-d1', 'PegTransferTBOTwoBlock-1000demo-2e5_0-d2', 'PegTransferTBOTwoBlock-1000demo-2e5_0-d3', 'Two blocks'),
#     ('PegTransferTBO-1000demo-2e5_0', 'PegTransfer-targetblock-1000demo-2e5_0-d1', 'PegTransfer-targetblock-1000demo-2e5_0-d2', 'Four blocks')
    
    ## TWO BLOCKS - done
#     ('PegTransferTwoBlocksNoColor-onehot-1000demo-2e5_0_d1', 'PegTransferTwoBlocksNoColor-onehot-1000demo-2e5_0_d2', 'PegTransferTwoBlocksNoColor-onehot-1000demo-2e5_0_d3', '⁠One hot block encoding'),
#     ('PegTransferTwoBlocksNoColor-onehottargetpeg-1000demo-2e5_0_fixed_r0', 'PegTransferTwoBlocksNoColor-onehottargetpeg-1000demo-2e5_0_fixed_r1', 'PegTransferTwoBlocksNoColor-onehottargetpeg-1000demo-2e5_0_fixed_r2', 'One hot block encoding and target peg position'),
#     ('PegTransferTwoBlocksNoColor-fourtuple-1000demo-2e5_0_fixed_r0', 'PegTransferTwoBlocksNoColor-fourtuple-1000demo-2e5_0_fixed_r1', 'PegTransferTwoBlocksNoColor-fourtuple-1000demo-2e5_0_fixed_r2', '⁠RGBA color'), 
#     ('PegTransferTwoBlocksNoColor-targetblock-1000demo-2e5_0_fixed_r0', 'PegTransferTwoBlocksNoColor-targetblock-1000demo-2e5_0_fixed_r1', 'PegTransferTwoBlocksNoColor-targetblock-1000demo-2e5_0_fixed_r2', 'Goal block position'),
#     ('PegTransferTwoBlocksNoColor-targetblocktargetpeg-1000demo-2e5_0_fixed_r0', 'PegTransferTwoBlocksNoColor-targetblocktargetpeg-1000demo-2e5_0_fixed_r1', 'PegTransferTwoBlocksNoColor-targetblocktargetpeg-1000demo-2e5_0_fixed_r2', "Goal block and target peg positions"),
    
    ## TWO BLOCKS COLOR - done
    ('PegTransferTwoBlocks-onehot-1000demo-2e5_0_fixed_r0', 'PegTransferTwoBlocks-onehot-1000demo-2e5_0_fixed_r1', '', '⁠One hot block encoding'),
    ('PegTransferTwoBlocksColor-onehottargetpeg-1000demo-2e5_fixed_r0', 'PegTransferTwoBlocksColor-onehottargetpeg-1000demo-2e5_fixed_r1', 'PegTransferTwoBlocksColor-onehottargetpeg-1000demo-2e5_fixed_r2', 'One hot block encoding and target peg position'),
    ('PegTransferTwoBlocks-fourtuple-1000demo-2e5_0_fixed_r0', 'PegTransferTwoBlocks-fourtuple-1000demo-2e5_0_fixed_r1', 'PegTransferTwoBlocks-fourtuple-1000demo-2e5_0_fixed_r2', "⁠RGBA color"),
    ('PegTransferTwoBlocksColor-targetblock-1000demo-2e5_0_fixed_r0', 'PegTransferTwoBlocksColor-targetblock-1000demo-2e5_0_fixed_r1', 'PegTransferTwoBlocksColor-targetblock-1000demo-2e5_0_fixed_r2', 'Goal block position'),
    ('PegTransferTwoBlocksColor-targetblocktargetpeg-1000demo-2e5_0_fixed_r0', 'PegTransferTwoBlocksColor-targetblocktargetpeg-1000demo-2e5_0_fixed_r1', 'PegTransferTwoBlocksColor-targetblocktargetpeg-1000demo-2e5_0_fixed_r2', "Goal block and target peg positions"),
    
    ## 50 epochs 1k vs. 10k demos - done
#     ('PegTransfer-targetblock-1000demo-1e5_0', 'PegTransfer-targetblock-10000demo-1e5_0-d2', 'PegTransfer-targetblock-10000demo-1e5_0-d1', '1k demos'),
#     ('PegTransfer-targetblock-10000demo-1e5_0', 'PegTransfer-targetblock-1000demo-1e5_0-d1', 'PegTransfer-targetblock-1000demo-1e5_0-d2', '10k demos')
]

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data_dict = {}

for exp_group in experiment_names:
    exp_data = []
    for experiment in exp_group[:3]:  # First three items are experiment names
        file_path = f"Downloads\\results\\{experiment}\\progress.csv"  # Update paths if needed
        print(file_path)
        try:
            exp_data.append(pd.read_csv(file_path))
        except FileNotFoundError:
            try:
                exp_data.append(pd.read_csv(f"Downloads\\logs\\logs\\her\\{experiment}\\progress.csv"))
            except FileNotFoundError:
                try:
                    exp_data.append(pd.read_csv(f"Downloads\\results\\060325_fixes\\{experiment}\\progress.csv"))
                except FileNotFoundError:
                    print(f"Warning: {file_path} not found, skipping.")

    if exp_data:
        data_dict[exp_group] = exp_data

if not data_dict:
    print("No valid data found. Please check file paths.")
    exit()

plt.figure(figsize=(10, 5))

for exp_group in data_dict:
    exp_data = data_dict[exp_group]
    print(exp_group)
    
    # Assume epochs are the same across experiments
    epochs = exp_data[0]["epoch"]

    success_rates = np.array([df["test/success_rate"].to_numpy() for df in exp_data])
    mean_success_rate = np.mean(success_rates, axis=0)
    std_success_rate = np.std(success_rates, axis=0)

    # Plot mean success rate with shaded standard deviation
    plt.plot(epochs, mean_success_rate, label=exp_group[3])
    plt.fill_between(epochs, mean_success_rate - std_success_rate, mean_success_rate + std_success_rate, alpha=0.2)

plt.xlabel("Epochs")
plt.ylabel("Evaluation Success Rate")
plt.title("Peg Transfer with Two Blocks (Observations with Color)")   #"Peg Transfer with Two Blocks" (Observations with Goal Block Position Only)
plt.legend(loc="upper left")
plt.grid(True)

plt.savefig("peg_transfer_two_blocks_color_experiments.jpg", bbox_inches='tight') # peg_transfer_tbo_experiments.jpg
plt.show()
```
